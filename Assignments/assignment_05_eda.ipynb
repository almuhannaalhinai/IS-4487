## 4. Univariate Analysis

Business framing:  

Hotels care about trends like average stay length, customer mix, and pricing. A good EDA starts with understanding single variables and their distribution.

- Select at least 3 individual variables to explore
- Use plots and summary methods (e.g. info(), describe(), etc)  to describe the distribution (hint: we are only looking at the values of one variable, so think of plots you've used in the past that DON'T compare 2 variables.)
- Focus on what matters from a business standpoint (e.g., pricing, cancellations, guest types)

```python
# Bar plot of cancellation status
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='is_canceled')
plt.title('Count of Canceled vs. Not Canceled Bookings')
plt.xlabel('Is Canceled (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])
plt.show()
```

```python
# Box plot for Average Daily Rate (adr)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[df['adr'] < 2000], y='adr')
plt.title('Distribution of Average Daily Rate (ADR) (ADR < 2000)')
plt.ylabel('ADR')
plt.show()

# Describe the adr column
display(df['adr'].describe())
```

```python
# Histogram for Lead Time
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='lead_time', bins=50, kde=True)
plt.title('Distribution of Lead Time')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Count')
plt.show()
```

### ‚úçÔ∏è Your Response: üîß
- **Is Canceled ‚Äì** I explored the is_canceled variable, which shows whether a booking was canceled or not. The bar chart reveals that about 37% of bookings were canceled, which is a significant portion and highlights cancellation as a serious business issue.

- **Average Daily Rate (ADR) ‚Äì** I explored the adr variable, which represents the price per night charged to customers. The boxplot and summary statistics indicate that most daily rates cluster between 50 and 150; however, there are extreme outliers, including negative and very high values, which suggest data quality issues that could distort pricing insights.

- **Lead Time ‚Äì** I explored the lead_time variable, which indicates how far in advance bookings were made. The histogram shows a strong right-skewed distribution, with many bookings made very close to the arrival date but some stretching out to nearly two years, suggesting hotels must prepare for both last-minute and long-term booking behaviors.


## 5. Bivariate Analysis

Business framing:  

Stakeholders often ask: ‚ÄúWhat drives cancellations?‚Äù or ‚ÄúDo longer stays mean higher revenue?‚Äù Bivariate analysis helps you uncover those kinds of relationships.

- Choose 2 relevant variable pairs (e.g., `lead_time` vs. `is_canceled`, or `adr` vs. `customer_type`)
- Use scatterplots, grouped bar plots, or boxplots to explore the relationships
- Interpret what these relationships could mean for the hotel business

```python
# Boxplot of lead_time split by is_canceled
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='is_canceled', y='lead_time')
plt.title('Lead Time vs. Cancellation Status')
plt.xlabel('Is Canceled (0 = No, 1 = Yes)')
plt.ylabel('Lead Time (Days)')
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='customer_type', y='adr', errorbar='sd')
plt.title('Average Daily Rate (ADR) by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()
```

The boxplot shows that cancellations generally have much longer lead times, meaning customers who book far in advance are more likely to cancel, which suggests the hotel should consider deposits or stricter policies for long-term reservations.

- **Average Daily Rate (ADR) vs. Customer Type:**  I analyzed how room pricing (ADR) varies by customer type.
